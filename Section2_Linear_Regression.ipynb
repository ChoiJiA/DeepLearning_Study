## Linear Regression(선형 회귀)
* 학습 Data : X(기본적인 자료, feature), Y(정답 값)
* Linear Hyphothesis : 가설에 맞는 선을 찾는 것(학습을 하기 위한)
* Hyphothesis 수식
    H(x) = Wx + b
    (H(x):가설, W:가중치, x(값), b(편향))
* Cost Function(손실함수) 
    손실 차이 : (H(x) - y)^2 
          -> 제곱하는 이유 : 양/음수의 영향을 받지 않고 차이를 극대화 할 수 있기 때문
    
    -> Cost Function의 목표 : 손실을 최소화, minimize cost(W, b) 

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
tf.set_random_seed(777) 
```

```python
# X/Y data
x_train = [1, 2, 3]
y_train = [1, 2, 3]
```

```python
# Try to find values for W and b to compute y_data = x_data * W + b
W = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
# hyphothesis : XW + b
hyphothesis = x_train * W + b
```

```python
# cost function
cost = tf.reduce_mean(tf.square(hyphothesis - y_train)) 
```

```python
# optimizer
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
# Launch the graph in a session.
with tf.compat.v1.Session() as sess:
    sess.run(tf.global_variables_initializer())

    # Fit the line
    for step in range(2001):
        _, cost_val, W_val, b_val = sess.run([train, cost, W, b])

        if step % 20 == 0:
            print(step, cost_val, W_val, b_val)
```
